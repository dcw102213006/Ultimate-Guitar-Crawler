```python
db=MongoConnect.mongoconnect()
one_document=db.Ultimate_Guitar_EachTAB_Html_byArtists
cursor=one_document.find({'error':0},no_cursor_timeout=True)
    
```

以下把Html Parse出Tab的資訊，包刮:歌手、譜名、版本、譜的所有和弦、譜的rating <br>
約32244筆資料，一首歌很多譜，32244筆資料要看有幾首歌

```python
do_data=0
with cursor:#做完後把這個non-timeout cursor清掉
    for i in cursor:
        try:
            do_data += 1 #處理到第幾筆
            if do_data % 100==0:{
                print('已完成',do_data,'筆;約完成',round(do_data/32244,3)*100,'%')
            }
            tab_href=i['tab_href']
            web_html=i['web_html']
            webhtml=i['web_html']
            
            print(tab_href)
            soup1=BeautifulSoup(web_html,'lxml')
            chord_list=soup1.find_all('span','_3L0Da _3f5kI')
            chord_dict=[]
            for chord in chord_list:
                chord_dict.append(chord.text)

            artist=soup1.find('span',class_='BOk-5').text
            #一頁html會列出所有version，要選有反白顏色的class
            version=soup1.find('a',class_='_1yyHg').find('span',class_='_25WMZ').text
            rating=soup1.find('a',class_='_1yyHg').find('div',class_='_36H3v').text
            tab_title=soup1.find('div',class_='_2Pu2_').text
            bag={
                'tab_href':tab_href,

                'artist':artist,
                'tab_title':tab_title,
                'version':version,
                'rating':rating,
                'chord':chord_dict,
                'error':0

            }

        except Exception as e:
            print('Parse這張譜時有問題',e)
            bag={
                'tab_href':tab_href,
                'web_html':'',
                'artist':'',
                'tab_title':'',
                'version':'',
                'rating':'',
                'chord':'',
                'error':1

            }
        finally:###連接MONGO




            db=MongoConnect.mongoconnect()
            one_document=db.Ultimate_Guitar_Tab_data_manyArtists
            current_insert=one_document.find({'tab_href':tab_href}).count()

            if(current_insert==0):
                one_document.insert_one(bag)
                print('insert success:',tab_href)
                
            if(current_insert>0):print("已存在此頁面")

```
